## 訓練出model

```python
from sklearn.datasets import load_iris
import lightgbm as lgb
import pandas as pd

iris = load_iris()

x = pd.DataFrame(iris['data'][:, 2:], columns=iris.feature_names[2:])
y = pd.DataFrame(iris['target'])

lgb_train = lgb.Dataset(x, y)

params={'objective':'multiclassova', 'num_class':3}

model = lgb.train(params, lgb_train)
```

## 畫圖

```python
import matplotlib.pyplot as plt

lgb.plot_tree(model, dpi=600,
              show_info=['split_gain', 'internal_value', 'internal_count', 'internal_weight',
                         'leaf_count', 'leaf_weight', 'data_percentage'])
plt.savefig('圖/lightgbm_tree.png')
plt.show()
```
