## 訓練出model

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()

x = iris['data'][:, 2:]
y = iris['target']

model = DecisionTreeClassifier(max_depth=2, criterion='entropy')
model.fit(x, y)
```

## 畫圖

### 將model輸出成dot檔

```python
from sklearn.tree import export_graphviz

export_graphviz(model, # 模型 
                out_file='圖/sklearn_tree.dot', # 輸出到dot檔
                feature_names=iris.feature_names[2:], # 顯示Feature Names
                class_names=iris.target_names, # 顯示target Name
                rounded=True, # 開啟round
                proportion=False, # 不顯示比例 顯示target的數量
                precision=2, # 小數點後第二位
                filled=True # 是否根據Target來填顏色
               )
```

###  用Jupyter notebook在終端機中將dot檔轉成png檔

```python
from subprocess import call
call(['dot', '-Tpng', '圖/sklearn_tree.dot', '-o', '圖/sklearn_tree.png'])
```
